# Jupyter Notebook for plotting the historical price data for a cryptocurrency - fiat currency

```python
import pandas as pd
import requests
import numpy as np
from datetime import datetime
from pycoingecko import CoinGeckoAPI
import plotly.express as px
```

## Hardcoding the parameters for now, will be making them as input parameters in the function definition

```python
coin = 'bitcoin'
curr = 'aud'
gecko = CoinGeckoAPI()
```

## Using the get_coin_market_chart_by_id() of the pycoingecko API object

```python
data = gecko.get_coin_market_chart_by_id( coin, vs_currency= curr, days= 60)
```

## Initialising and setting the dataframe

```python
df_data = pd.DataFrame(columns=['date', 'close'], index=None)

for i in data['prices']:

    df_data = df_data.append(pd.Series([
        i[0],
        i[1]
    ], index=df_data.columns), ignore_index= True)

df_data
```

```python
# Converting from the unix timestamp 
df_data['date'] = pd.to_datetime( df_data['date'], unit='ms', origin='unix' ) 
df_data
```

```python
fig = px.line(df_data, x='date', y="close")
fig.show()
```
